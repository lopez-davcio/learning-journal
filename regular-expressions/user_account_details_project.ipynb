# <center>User Account Details Project<center>

This notebook focuses on validating user account information according to strict formatting and security rules. Each entry (first name, last name, date of birth, email address, username, password, and credit card information), must meet precise criteria to ensure data accuracy, consistency, and security.  
The goal is to simulate a real-world account creation process where every input is thoroughly validated before acceptance.

---

```python
import re
```

### First Name:
It does not contain any other characters besides letters.  
It starts with a capital letter and all the other letters are lowercase.  
It is at least 2 characters long.

```python
while True:
    fname = input("\nPlease enter your First Name: ")

    check = re.fullmatch(r"^[A-Z][a-z]+", fname)

    if check == None:
        print("Wrong format! Please try again.")
        continue
    else:
        break
```

### Date of birth:
The person was born before January 1st 2002.  
It contains only digits and forward slashes.  
Follows the format MM/DD/YYYY.  

```python
import re
while True:
    date = input("\nPlease enter your Date of Birth (mm/dd/yyyy): ")
    
    check = re.fullmatch(r"(?:0[1-9]|1[0-2])/(?:0[1-9]|[12][0-9]|3[0-1])/(?:19\d{2}|200[01])", date)

    if check == None:
        print("Wrong format! Please try again.")
        continue
    else:
        break
```

---

### Email address:

Follows the correct format: `username@domain.extension`.  
The username may contain only letters, digits, dots or the underscore character.  
The domain may contain only lowercase letters.  
The extension may contain only 2, 3 or 4 lowercase letters.

```python
while True:
    email = input("\nPlease enter your Email Address: ")
    
    check = re.fullmatch(r"[\w.]+@[a-z]+\.[a-z]{2,4}", email)

    if check == None:
        print("Wrong format! Please try again.")
        continue
    else:
        break
```

---

### Username:

It does not contain any other characters besides letters, digits or the underscore character.  
Has a minimum of 6 characters and a maximum of 12 characters.

```python
while True:
    user = input("\nPlease enter your Username: ")
    
    check = re.fullmatch(r"\w{6,12}", user)

    if check == None:
        print("Wrong format! Please try again.")
        continue
    else:
        break
```

---

### Password:  
It's at least 8 characters long.  
It starts with a lowercase letter.  
It should contain at least one lowercase letter, one uppercase letter, one digit and one of the following characters: $, &, ?, ! or %.

```python
while True:
    passw = input("\nPlease enter your Password: ")
    
    check = re.fullmatch(r"(?=.*[$&?!%])(?=.*[A-Z])(?=.*\d)[a-z].{7,}", passw)
    
    if check == None:
        print("Wrong format! Please try again.")
        continue
    else:
        break
```

---

### Credit card number:  
The card should be issued by Visa or Mastercard only (starting with a 4 or a 5).  
It is exactly 16 digits long and contains no other characters besides digits.

```python
while True:
    ccnum = input("\nPlease enter your Credit Card Number (no spaces): ")

    check = re.fullmatch(r"[45]\d{15}", ccnum)

    if check == None:
        print("Wrong format! Please try again.")
        continue
    else:
        break
```

---

### Credit card expiration date:  
It's not sooner than May 2025 inclusively.  
It contains only digits and forward slashes.  
It follows the format MM/YY.

```python
while True:
    ccdat = input("\nPlease enter your Credit Card Expiration Date (mm/yy): ")
    
    check = re.fullmatch(r"(?:0[5-9]/25)|(?:1[012]/25)|(?:0[1-9]/2[5-9])|(?:1[012]/2[5-9])|(?:0[1-9]/3\d)|(?:1[012]/3\d)", ccdat)
    
    if check == None:
        print("Wrong format! Please try again.")
        continue
    else:
        break
```

---

### Credit card verification code / verification value (CVC / CVV):  
It's a 3 digit number.  
It does not contain any other characters.

```python
while True:
    cccvc = input("\nPlease enter your Credit Card Verification Code: ")
    
    check = re.fullmatch(r"\d{3}", cccvc)
    
    if check == None:
        print("Wrong format! Please try again.")
        continue
    else:
        break
```
