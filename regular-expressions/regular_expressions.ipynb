### Exercises from **bookshelf.txt**

The data in bookshelf.txt is structured as semicolon-separated values, with each line following the format:

*Author Name;Book Title;Year of Publication*  

Example:
>*Charlie Chaplin;My Autobiography;1964*

```python
import re

with open('bookshelf.txt', 'r') as f:
    bookshelf = f.read()
```

---

Match all book titles ending with the letter p.

```python
re.findall(r".+?;(.+p);.+?", bookshelf)
```

---

Match all the authors in the file whose last name starts with the letter B.

```python
matches = re.findall(r"^\w+? B\w+(?=;)", bookshelf, re.M)
print(matches[:9])
```

---

Match all the books in the file that have been published between 1980 and 1999.

```python
matches = re.findall(r".+;(.+?);19[8-9][0-9]", bookshelf)
print(matches[:7])
```

### Exercises from **phone.txt**

The data in phone.txt is structured as a tab-separated line, with each line following the format:

*Full Name&nbsp;&nbsp;&nbsp;&nbsp;Phone Number*
Example:
>*Oliver Cooper&nbsp;&nbsp;&nbsp;&nbsp;(917) 965-9749*

```python
with open('phone.txt', 'r') as f:
    phone = f.read()
```

---

Match the last name and phone number of all entries having an area code ending in zero.

```python
matches = re.findall(r' (.+)\t(\(\d{2}0\).+)', phone)
print(matches[:9])
```

---

Match the area code of all phone numbers ending in 7.

```python
matches = re.findall(r'\(\d{3}\)(?= \d{3}-\d{3}7)', phone)
print(matches[:9])
```

---

Match first and last name of those with a phone number starting with an odd digit 

```python
matches = re.findall(r'^.+(?=\t.{6}[13579])', phone, re.M)
print(matches[:9])
```

---

Match first name of those whose area code is a number less than 300.

```python
matches = re.findall(r'^\w+\b(?=.+\([012])', phone, re.M)
print(matches[:9])
```

---

Match first name of those whose last name ends in a vowel and phone number ends in either 0, 7 or 9.

```python
re.findall(r'^(\w+) \w+[aeiou]\b.+[079]$', phone, re.M)
```

### Exercises from **logs.txt**

The data in logs.txt is structured as a space-separated line, with each line following the format:  
**Level Date and Time Source Event ID Task Category**

  
Example:   
>*Critical 1/27/2020 8:19:22 AM Microsoft-Windows-Kernel-Power 41 (63) The system has rebooted without cleanly shutting down first. This error could be caused if the system stopped responding, crashed, or lost power unexpectedly.*

```python
with open('logs.txt', 'r') as f:
    logs = f.read()
```

---

Extract the source substring of all critical logs generated between the 11th and 16th of January 2020.

```python
matches = re.findall(r'^Critical 1/1[1-6]/2020 .+[AP]M (.+?)\b\d', logs, re.M)
print(matches)
```

---

Extract the source substring of all log entries that were generated after 12 PM and before 4 PM.

```python
re.findall(r'.+2020 (?:12|1|2|3):\d\d:\d\d PM (.+?) \d', logs)
```

---

Match the Date of all the log entries that have TPM as the Source of the log message.

```python
matches = re.findall(r"(\d\d*/\d{2}/\d{4})\b.+TPM", logs)
print(matches[:9])
```

---

Match the Date and Time of all the log entries that have been generated between 24 and 27 of January 2020 and between 8:00:00 and 8:59:59 AM.

```python
re.findall(r'1/2[4-7]/2020 8:\d{2}:\d{2}',logs)
```
